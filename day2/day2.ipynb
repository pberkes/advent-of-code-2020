```python
from collections import Counter
```

```python
with open('input.txt', 'r') as f:
    passwords = f.readlines()
```

```python
passwords[:2]
```

```python
def parse_password(password):
    policy, pwd = password.strip().split(':')
    pwd = pwd.strip()
    range_, letter = policy.split(' ')
    min_, max_ = range_.split('-')
    return letter, int(min_), int(max_), pwd
```

```python
def is_valid(password):
    letter, min_, max_, pwd = parse_password(password)
    n_letter = Counter(pwd)[letter]
    valid = min_ <= n_letter <= max_
    return valid
```

```python
cases = [
    ('1-3 a: abcde', True),
    ('1-3 b: cdefg', False),
    ('2-9 c: ccccccccc', True),
] 

for p, expected in cases:
    print(is_valid(p) == expected)
```

```python
n_valid = 0
for p in passwords:
    print(p.strip(), is_valid(p))
    n_valid += is_valid(p)
```

```python
def is_valid_2(password):
    letter, pos1, pos2, pwd = parse_password(password)
    valid = ((pwd[pos1 - 1] == letter) + (pwd[pos2 - 1] == letter)) == 1
    return valid
```

```python
cases = [
    ('1-3 a: abcde', True),
    ('1-3 b: cdefg', False),
    ('2-9 c: ccccccccc', False),
] 

for p, expected in cases:
    print(is_valid_2(p) == expected)
```

```python
n_valid = 0
for p in passwords:
    print(p.strip(), is_valid_2(p))
    n_valid += is_valid_2(p)
```
