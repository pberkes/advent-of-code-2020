```python
test_map = """
..##.......
#...#...#..
.#....#..#.
..#.#...#.#
.#...##..#.
..#.##.....
.#.#.#....#
.#........#
#.##...#...
#...##....#
.#..#...#.#
""".strip()

expected = 7
```

```python
def trees_hit(map_, pattern=(3, 1)):
    pos = [0, 0]
    map_ = [list(row) for row in list(map_.split('\n'))]
    map_height = len(map_)
    map_width = len(map_[0])
    
    hits = 0
    while pos[1] < map_height:
        if map_[pos[1]][pos[0]] == '#':
            hits += 1
        pos[0] = (pos[0] + pattern[0]) % map_width
        pos[1] += pattern[1]
    return hits
```

```python
assert trees_hit(test_map) == expected
```

```python
cases = [
    ((1, 1), 2),
    ((3, 1), 7),
    ((5, 1), 3),
    ((7, 1), 4),
    ((1, 2), 2),
]

tot = 1
for pattern, expected_hits in cases:
    hits = trees_hit(test_map, pattern)
    assert hits == expected_hits
    tot *= hits
    
assert tot == 336

```
